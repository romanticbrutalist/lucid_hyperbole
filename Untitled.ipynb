```python
def vid_split(vid_dir,tempdir,t_img_format,file_prefix='frame'):
    
    if not os.path.exists(tempdir):
        os.mkdir(tempdir)
    if not os.path.exists(tempdir+"org/"):
        os.mkdir(tempdir+"org/")
    if not os.path.exists(tempdir+"r/"):
        os.mkdir(tempdir+"r/")
    vidcap = cv2.VideoCapture(vid_dir)
    #vidcap.set(cv2.CAP_PROP_BRIGHTNESS, 1)
    vidcap.set(cv2.CAP_PROP_SATURATION, 0.5)
    success,image = vidcap.read()
    count = 0
    while success:
        cv2.imwrite("{1}org/{2}{0}.{3}".format(count,tempdir,file_prefix,t_img_format), image)     # save frame as JPEG file      
        success,image = vidcap.read()
        
        #print('Read a new frame: ', success)
        count += 1
        #print(img)
    
    simg = cv2.imread('{1}org/{2}{0}.{3}'.format(0,tempdir,file_prefix,t_img_format))
    height, width, layers = simg.shape
    size = (width,height)
```

```python
import cv2
import numpy as np
import glob
import os
import shutil

def rotateImage(image, angle):
  image_center = tuple(np.array(image.shape[1::-1]) / 2)
  rot_mat = cv2.getRotationMatrix2D(image_center, angle, 1.0)
  result = cv2.warpAffine(image, rot_mat, image.shape[1::-1], flags=cv2.INTER_LINEAR)
  return result
def contVidRotate(vid_dir,out_name,dpf,wr_format,fcc,out_dir,rotate_time_start_perc,tempdir,t_img_format,file_prefix='frame'):
    
    if not os.path.exists(tempdir):
        os.mkdir(tempdir)
    if not os.path.exists(tempdir+"org/"):
        os.mkdir(tempdir+"org/")
    if not os.path.exists(tempdir+"r/"):
        os.mkdir(tempdir+"r/")
    vidcap = cv2.VideoCapture(vid_dir)

    success,image = vidcap.read()
    count = 0
    while success:
        cv2.imwrite("{1}org/{2}{0}.{3}".format(count,tempdir,file_prefix,t_img_format), cv2.convertScaleAbs(image, alpha=1.25, beta=-24))     # save frame as JPEG file      
        success,image = vidcap.read()
        #print('Read a new frame: ', success)
        count += 1
    print("c0")
    simg = cv2.imread('{1}org/{2}{0}.{3}'.format(0,tempdir,file_prefix,t_img_format))
    height, width, layers = simg.shape
    size = (width,height)
    l = glob.glob('{0}/org/{1}*.jpg'.format(tempdir,file_prefix))
    rs = len(l)*rotate_time_start_perc
    for i in range(len(l)):
        img0 = cv2.imread('{1}org/{2}{0}.{3}'.format(i,tempdir,file_prefix,t_img_format))
        
        if i>rs:
            img0 = rotateImage(img0,(i-rs)*dpf)
    
        cv2.imwrite('{1}r/{2}{0}.{3}'.format(i,tempdir,file_prefix,t_img_format), img0)
    print("c1")
    img_array = []

    for i in range(len(l)):
        img = cv2.imread('{1}r/{2}{0}.{3}'.format(i,tempdir,file_prefix,t_img_format))
        height, width, layers = img.shape
        size = (width,height)
        img_array.append(img)
 
    print(len(img_array),size)
    ofd = '{0}{1}.{2}'.format(out_dir,out_name,wr_format)
    out = cv2.VideoWriter('{0}{1}.{2}'.format(out_dir,out_name,wr_format),cv2.VideoWriter_fourcc(*'MP4V'), 30, size)
    print(ofd)
    for i in range(len(img_array)):
        out.write(img_array[i])
    out.release()
    #shutil.rmtree(tempdir)
    
# True
```
